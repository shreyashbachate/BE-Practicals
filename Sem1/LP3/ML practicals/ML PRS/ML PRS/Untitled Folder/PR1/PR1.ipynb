```python
# Function to calculate distance
from math import *

def distance_formula(longitude1,latitude1,longitude2,latitude2):
    travel_dist = []
    for pos in range(len(longitude1)):
        lon1,lan1,lon2,lan2 = map(radians,[longitude1[pos],latitude1[pos],longitude2[pos],latitude2[pos]])
        dist_lon = lon2 - lon1
        dist_lan = lan2 - lan1
        
        a = sin(dist_lan/2)**2 + cos(lan1) * cos(lan2) *sin(dist_lon/2)**2
        
#         radius of earth = 6371
        c = 2*asin(sqrt(a))*6371
        travel_dist.append(c)
    return travel_dist
```

```python
df['dist_travel_km'] = distance_formula(df.pickup_longitude.to_numpy(),df.pickup_latitude.to_numpy(),df.dropoff_longitude.to_numpy(),df.dropoff_longitude.to_numpy())
```

```python
# Identify Outliers
```

```python
df.plot(kind = 'box',subplots = True, layout = (6,2),figsize = (15,20))
plt.show()
```

```python
# Use IQR to fill values
def remove_outliers(df1,col):
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3-Q1
    low = Q1 - 1.5*IQR
    high = Q3 + 1.5*IQR
    df[col] = np.clip(df1[col],low,high)
    return df1

def treat_outliers_all(df1,col_list):
    for c in col_list:
        df1 = remove_outliers(df,c)
    return df1
```

```python
df = treat_outliers_all(df,df.iloc[:0,0::])
```

```python
# Linear Regression and Random Forest
```

```python
x = df.drop('fare_amount',axis = 1)
y = df['fare_amount']
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(x_train,y_train)
```

```python
y_pred_lin = reg.predict(x_test)
print(y_pred_lin)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators = 100)
model.fit(x_train,y_train)
```

```python
y_pred_rf = model.predict(x_test)
print(y_pred_rf)
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
rmse_lin = np.sqrt(metrics.mean_squared_error(y_test,y_pred_lin))
r2_lin = r2_score(y_test,y_pred_lin)
ans = pd.Series({'Model': 'Linear Regression', 'RMSE':rmse_lin , 'R2': r2_lin})
print(ans)
```

```python
rmse_lin = np.sqrt(metrics.mean_squared_error(y_test,y_pred_rf))
r2_lin = r2_score(y_test,y_pred_rf)
ans = pd.Series({'Model': 'Random Forest', 'RMSE':rmse_lin , 'R2': r2_lin})
print(ans)
```
