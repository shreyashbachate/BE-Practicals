```python
df = df.drop(['RowNumber','CustomerId','Surname'],axis = 1)
```

```python
df
```

```python
geography = pd.get_dummies(df['Geography'])
```

```python
gender = pd.get_dummies(df['Gender'])
```

```python
df = pd.concat([df,geography,gender],axis = 1)
```

```python
df
```

```python
df = df.drop(['Geography','Gender'],axis = 1)
```

```python
df
```

```python
# Split the data
x = df.drop('Exited',axis = 1)
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = 0)
```

```python
# Fearture Scaling
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
f
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(units = 6,kernel_initializer = 'he_uniform',activation = 'relu',input_dim = 13))
```

```python
classifier.add(Dense(units = 6,kernel_initializer = 'he_uniform',activation = 'relu'))
```

```python
classifier.add(Dense(units = 1,kernel_initializer = 'glorot_uniform',activation = 'sigmoid'))
```

```python
classifier.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
model_history = classifier.fit(x_train,y_train,batch_size = 10,validation_split = 0.33,epochs = 100)
```

```python
model_history.history.keys()
```

```python
plt.figure(figsize = (6,6))
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
```

```python
plt.figure(figsize = (6,6))
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
```

```python
# Prediction
y_pred = classifier.predict(x_test) 
```

```python
y_pred
```

```python
y_pred = y_pred>0.5
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
asco = accuracy_score(y_test,y_pred)
asco
```
