```python
import pandas as pd
import numpy as np
import sympy as sym
import matplotlib
from matplotlib import pyplot
```

```python
def objective(x):
    return (x+3)**2
```

```python
def derivative(x):
    return 2*(x+3)
```

```python
def gradient(alpha,start,max_itr):
    x_list = list()
    x = start
    x_list.append(x)
    for i in range(max_itr):
        gradi = derivative(x);
        x = x - (alpha * gradi)
        x_list.append(x)
    return x_list
```

```python
x = sym.symbols('x')
expr = (x+3)**2.0
gradi = sym.Derivative(expr,x)
print(gradi.doit())
print(gradi.doit().subs(x,2))
```

```python
alpha = 0.1
start = 2
max_itr = 30
```

```python
x_cor = np.linspace(-15,15,100)
pyplot.plot(x_cor,objective(x_cor))
pyplot.plot(2,objective(2),'ro')
```

```python
x = gradient(alpha,start,max_itr)
local_minimum = x[-1]
x_cor = np.linspace(-5,5,100)
pyplot.plot(x_cor,objective(x_cor))
x_arr = np.array(x)

pyplot.plot(local_minimum,objective(local_minimum),'go')
pyplot.plot(x_arr,objective(x_arr),'.-',color ='red')
```
