```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
scaled_data=sc.fit_transform(df[['QUANTITYORDERED','SALES']])
```

```python
wcss = []
```

```python
from sklearn.cluster import KMeans
# Trying k values from 1 to 10
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,random_state=0)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.plot(range(1,11),wcss,marker='o',linestyle='-',color='b')
plt.title("Elbow method")
plt.xlabel("No of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
optimal_k = 3
```

```python
kmeans=KMeans(n_clusters=3,random_state=0)
kmeans.fit(scaled_data)
df = df.assign(Cluster = kmeans.labels_)
```

```python
df['Cluster']
```

```python
df.head()
```

```python
plt.figure(figsize = (9,6))
for cluster in range(optimal_k):
    plt.scatter(df[df['Cluster'] == cluster]['QUANTITYORDERED'],df[df['Cluster'] == cluster]['SALES'],label = f'cluster {cluster}')
plt.title(f'Kmeans Clustering with {optimal_k} clusters')
plt.xlabel('QUANTITYORDERED')
plt.ylabel('SALES')
plt.legend()
plt.show()
```
