```python
df = df.drop(['ADDRESSLINE2','STATE','POSTALCODE','TERRITORY'],axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
lc = LabelEncoder()
```

```python
df['STATUS'] = lc.fit_transform(df['STATUS'])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_data = sc.fit_transform(df[['QUANTITYORDERED','SALES']])
```

```python
wcss = []
```

```python
from sklearn.cluster import KMeans
for k in range(1,11):
    kmeans = KMeans(n_clusters =k,random_state = 0)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (9,6))
plt.plot(range(1,11),wcss,marker = 'o',linestyle = '-',color = 'b')
plt.title('Elbow Method')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
optimal_k = 3
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 0)
kmeans.fit(scaled_data)
df = df.assign(Clusters = kmeans.labels_)
```

```python
df['Clusters']
```

```python
df.head()
```

```python
plt.figure(figsize = (9,6))

for cluster in range (optimal_k):
    plt.scatter(df[df['Clusters'] == cluster]['QUANTITYORDERED'],df[df['Clusters'] == cluster]['SALES'],label = f'Cluster {cluster}')

plt.title(f'KMeans clustering with {optimal_k} clusters')
plt.xlabel('QUANTITYORDERED')
plt.ylabel('SALES')
plt.legend()
plt.show()
```
