```python
df = df.drop(['RowNumber','CustomerId','Surname'],axis = 1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (9,9))
sns.countplot(x = df['Exited'],data = df)
plt.xlabel("0: Not Exited  1:Exited")
plt.ylabel('count')
plt.show()
```

```python
geography = pd.get_dummies(df['Geography'])
```

```python
gender = pd.get_dummies(df['Gender'])
```

```python
df = pd.concat([df,geography,gender],axis = 1)
```

```python
df
```

```python
df = df.drop(['Geography','Gender'],axis = 1)
```

```python
df
```

```python
x = df.drop('Exited',axis = 1)
```

```python
y = df['Exited']
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim = 13))
```

```python
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
```

```python
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation = 'sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model_history = classifier.fit(x_train,y_train,batch_size = 10,validation_split=0.33,epochs=100)
```

```python
model_history.history.keys()
```

```python
plt.figure(figsize=(6,6))
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
```

```python
plt.figure(figsize=(6,6))
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
y_pred = y_pred > 0.5
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
asco = accuracy_score(y_test,y_pred)
asco
```
